## 1.3. Visualização dos dados em formato de gráficos

```python
sepal_length = X["sepal length"]
sepal_width = X["sepal width"]
petal_length = X["petal length"]
petal_width = X["petal width"]
classes = y["class"]
```

```python
np.unique(classes)
```

```python
uclasses = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
```

```python
mapa_cor = { 'Iris-setosa': "yellow", 'Iris-versicolor': "blue", 'Iris-virginica': "green" }
```

```python
cores = [ mapa_cor[c] for c in classes ]
```

### 1.3.1. sepal length vs sepal width


---





```python
plt.scatter(sepal_length[0:49], sepal_width[0:49], c=cores[0:49], alpha=0.5, label=uclasses[0])
plt.scatter(sepal_length[50:99], sepal_width[50:99], c=cores[50:99], alpha=0.5, label=uclasses[1])
plt.scatter(sepal_length[100:149], sepal_width[100:149], c=cores[100:149], alpha=0.5, label=uclasses[2])
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.legend(loc="upper right")
plt.show()
```

Observe que a espécie *Iris setosa* é linearmente separável das demais espécies: *Iris versicolor* e *Iris virginica* quando se analisa o comprimento e largura da sépala.

```python
plt.scatter(sepal_length[0:49], sepal_width[0:49], c=cores[0:49], alpha=0.5, label=uclasses[0])
plt.scatter(sepal_length[50:99], sepal_width[50:99], c=cores[50:99], alpha=0.5, label=uclasses[1])
plt.scatter(sepal_length[100:149], sepal_width[100:149], c=cores[100:149], alpha=0.5, label=uclasses[2])
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.legend(loc="upper right")
plt.plot([4.3, 6.8], [2, 4.5], c="black") #Uma Linha que separa o conjunto Iris setosa
plt.show()
```
