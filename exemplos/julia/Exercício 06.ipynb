```julia
using Printf
```

```julia
f(x) = -3x^6 + 2x^5 + 8x^2
```

```julia
g(x) = -18x^5 + 10x^4 + 16x
```

```julia
@printf("| %3s | %6s | %6s | %6s |\n", 
            "t", "x_t", "g(x_t)", "f(x_t)")

x = -0.3
for t = 1:40
    xa = x
    gxt = g(x)
    fxt = f(x)
    
    x = x + 0.005 * gxt
    
    @printf("| %3d | %6.2f | %6.3f | %6.3f |\n", 
        t, xa, gxt, fxt)
end
```

```julia
using Plots
```

```julia
xs = collect(-1.3:0.01:1.6);
```

```julia
ys = f.(xs);
```
